```python
import random
import matplotlib.pyplot as plt
import numpy as np
def lp(x,y,x0):
    size = len(x)
    y0=0
    for i in range(0,size):
        f=1
        for j in range(0,size):
            if i!=j:
                f=f *(x0-x[j])/(x[i]-x[j])
        y0+=f*y[i]
    return x0,y0
x,y = [],[]
fnY = []
while len(x) < 10:
    temp1 = random.randint(-1000,1000)
    if temp1 not in x :
        x.append(temp1)
x.sort()
while len(y) < 500:
    temp1 = random.randint(-1000,1000)
    if temp1 not in y :
        y.append(temp1)

        
xlin = np.linspace(x[0],x[-1],num =100)
for i in xlin :
    fx,fy = lp(x,y,i)
    fnY.append(fy)

    
```

```python
plt.title("Lagrange Polynomial-Prediction")
plt.plot(xlin,fnY)
```
